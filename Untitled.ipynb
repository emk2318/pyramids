```python
#to make things easier, we can put the input requirements into a variable called prompt
prompt = 'enter a number 1 to 8: ' #they know it's a string, because it's smart
number = int(input(prompt))
```

```python
#ok, now the next part I think is input validation, so let's make a variable for telling people they're dumb
error_message = 'seriously? can you not read the instructions?'
```

```python
#now, we're going to use a while loop to confirm the number is within the correct range
# to start this, let's say our number is 0, outside of the range
#if we didn't do this, the program wouldn't prompt for user input
number = 0
while number < 1 or number > 8:
    number = int(input(prompt))
    if number < 1 or number > 8:
        print(error_message)
```

```python
#ok so we're going to loop through all the numbers, adding rows each time, until we've reached the user's number
```

```python
#just saying here we're going to keep going from 0 to the number adding rows
for row in range(number): 
    #the number of periods in the pyramid is the user's number, minus the row number, minus 1.
    #the number of hashtags is the user's number minus the number of periods
    #so, the first row with a user number of 8 has 7 periods, 1 hashtag, and so on. 
    periods = (number - row) - 1 
    hashtags = number - periods
    #just saying below, multiply the period/hashtag string by the number of periods/hashtags and store them in variables
    period_string = '.' * periods 
    hashtags_string = '#' * hashtags
    print(period_string+hashtags_string)

    
```
